Project 4 : Data Visualization

Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Step 3: Load Dataset

```python
df = pd.read_excel("Cleaned_Dataset.xlsx")
```

Check Dataset

```python
df.head()
```

```python
df.shape
```

Bar Chart (Most Purchased Products)

```python
product_counts = df['Product'].value_counts()

product_counts.plot(kind='bar')

plt.title("Product Distribution")
plt.xlabel("Products")
plt.ylabel("Number of Orders")

plt.show()
```

Pie Chart (Payment Method)

```python
payment = df['PaymentMethod'].value_counts()

payment.plot(kind='pie', autopct='%1.1f%%')

plt.title("Payment Method Distribution")

plt.ylabel("")

plt.show()
```

Bar Chart (Order Status)

```python
status = df['OrderStatus'].value_counts()

status.plot(kind='bar')

plt.title("Order Status")

plt.xlabel("Status")

plt.ylabel("Orders")

plt.show()
```

Bar Chart (Referral Source)

```python
referral = df['ReferralSource'].value_counts()

referral.plot(kind='bar')

plt.title("Referral Sources")

plt.xlabel("Source")

plt.ylabel("Customers")

plt.show()
```

Monthly Sales Trend (Line Chart)

```python
df['Month'] = df['Date'].dt.to_period('M')
```

```python
monthly_sales = df.groupby('Month')['TotalPrice'].sum()

monthly_sales.plot()

plt.title("Monthly Sales Trend")

plt.xlabel("Month")

plt.ylabel("Sales")

plt.show()
```

Histogram (Total Price)

```python
df['TotalPrice'].plot(kind='hist', bins=20)

plt.title("Distribution of Total Price")

plt.xlabel("Total Price")

plt.show()
```
